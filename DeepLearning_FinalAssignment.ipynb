```python
from tensorflow import keras
import matplotlib.pyplot as plt
```

## Data preparation

 As the Kaggle entry states, the folder 'dataset2-master' contains 2,500 augmented images with 4 additional subtype labels (JPEG + CSV).

```python
# Make path variable and determine desired image dimensions
train_path = r'C:\\Users\\Martin Schepers\\Documents\dataset2-master\\dataset2-master\\images\\TRAIN'
imgDim = (120, 160, 3)

# the data has already been prepared, now it has to be loaded, split and shuffled
# For this we use a build-in function which creates a generator.
trainData = keras.utils.image_dataset_from_directory(
    train_path,
    shuffle=True,
    labels='inferred',
    label_mode='int',
    validation_split=0.25,
    subset="training",
    seed=42,
    image_size=imgDim[:2])

valData = keras.utils.image_dataset_from_directory(
    train_path,
    shuffle=True,
    labels='inferred',
    label_mode='int',
    validation_split=0.25,
    subset="validation",
    seed=42,
    image_size=imgDim[:2])
```

```python
for x_train, y_train in trainData:
    print(x_train.shape, x_train.dtype)
    print(y_train.shape, y_train.dtype)
    break
```

```python
import matplotlib.pyplot as plt
for i in range(0, 10):
    plt.imshow(x_train[i]/255)
    plt.title(str(y_train[i]))
    plt.show()
```

## Deep learning model creation

```python
#### Change to leaky relu! So far best
#### Perhaps higher dropout?

celclass = keras.models.Sequential()
# rescale images (divide by 255 to get all values in between 0 and 1)
celclass.add(keras.layers.Rescaling(1/255, input_shape=imgDim))
# First convolutional layer
celclass.add(keras.layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu'))
celclass.add(keras.layers.MaxPooling2D())
celclass.add(keras.layers.Dropout(0.3))

# Second convolutional layer
celclass.add(keras.layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
celclass.add(keras.layers.MaxPooling2D(pool_size = (2,2)))
celclass.add(keras.layers.Dropout(0.3))

# Third convolutional layer
celclass.add(keras.layers.Conv2D(filters=128, kernel_size=(3,3), activation='relu'))
celclass.add(keras.layers.MaxPooling2D())

# Dense layer
celclass.add(keras.layers.Flatten())
celclass.add(keras.layers.Dense(units=128, activation='relu'))
celclass.add(keras.layers.Dropout(0.3))

# Output layer
celclass.add(keras.layers.Dense(units=4, activation='softmax'))

celclass.summary()
```

```python
celclass.compile(
    loss = 'sparse_categorical_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy'])
```

```python
history = celclass.fit(trainData, validation_data=valData, epochs=30)
```

```python
history.history.keys()
```

```python
plt.plot(history.history['loss'], 'go', label='Training loss')
plt.plot(history.history['val_loss'], 'g-', label='Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['accuracy'], 'go', label='Categorical accuracy')
plt.plot(history.history['val_accuracy'], 'g-', label='Validation categorical accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
