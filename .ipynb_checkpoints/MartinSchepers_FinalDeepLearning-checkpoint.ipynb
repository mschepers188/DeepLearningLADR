 As the Kaggle entry states, the folder 'dataset2-master' contains 2,500 augmented images with 4 additional subtype labels (JPEG + CSV). Keras can easily load these images by creating a generator:

```python
# Make path variable and determine desired image dimensions
# test_path = r'/homes/maschepers/Documents/Cell_data_DeepLearning/dataset2-master/dataset2-master/images/TEST/'
# train_path = r'/homes/maschepers/Documents/Cell_data_DeepLearning/dataset2-master/dataset2-master/images/TRAIN/'
test_path = r'C:\\Users\\Martin Schepers\\Documents\\dataset2-master\\dataset2-master\\images\\TEST\\'
train_path = r'C:\\Users\\Martin Schepers\\Documents\\dataset2-master\\dataset2-master\\images\\TRAIN\\'


train_data = keras.preprocessing.image.ImageDataGenerator(rescale=1/255)
train_generator = train_data.flow_from_directory(
    directory=train_path,
    batch_size=20,
    color_mode='rgb',
    target_size=(120, 160),
    class_mode='categorical'
)

test_data = keras.preprocessing.image.ImageDataGenerator(rescale=1/255)
test_generator = test_data.flow_from_directory(
    directory=test_path,
    batch_size=20,
    color_mode='rgb',
    target_size=(120, 160),
    class_mode='categorical'
)
```

It seems the generator found about 10000 images for training and around 2500 for testing, both having a total of 4 classes. This is in accordance to kaggle as well as what is found checking the folder manually.


Keras can immmediately take the images from the generator and perform deep learning, but to test the baseline models the images need to be preprocessed. 

There is no Keras function to deliver the images to the baseline models, thus a function was written to extract the images. This subset will also be used for the fully connected model, as using the generator kept causing the computer to run out of memory (school "nuc")

```python
def get_data(generator, batch_size):
    # Store the data in X, y variables by iterating over the batches
    generator.reset()
    X, y = next(generator)
    for i in range(int(len(generator)/batch_size)-1): #1st batch is already fetched before the for loop.
        img, label = next(generator)
        X = np.append(X, img, axis=0 )
        y = np.append(y, label, axis=0)
    print(X.shape, y.shape)
    return X, y

# A subset of the data will be used as to not overload these models
X_train, y_train = get_data(train_generator, 2)
X_test, y_test = get_data(test_generator, 2)
```

```python
print("train", X_train.shape, y_train.shape)
print("test", X_test.shape, y_test.shape)
```

```python
plt.figure(figsize=(15, 5))
for i in range(0, 15):
    plt.subplot(3, 5, i+1)
    plt.imshow(X_train[i])
    plt.title(str(y_train[i]))
plt.show()
```

These images already seem to have been pre-processed as stated in kaggle, they all have the same dimensions and seem to have been turned a few degrees. They all contain stained cells, which will belong to one of the four classes.

The next step is to see if the data is properly distributed:

```python
# make array like labels into strings for plotting
trainlab = [str(i) for i in y_train]
testlab = [str(i) for i in y_test]

# plot histogram
plt.hist(trainlab, label='Training data')
plt.hist(testlab, label='Test data')
plt.title('Blood cells data distribution')
plt.legend()
plt.show()
```

The data seems to be very well distributed for both the training and test data. Each class is represented equally as can be seen in the histogram above.

### Baseline model

Fit and evaluate one or two classical (i.e. "not deep") supervised machine learning methods that are most suitable to serve as a reference to assess a baseline performance. Briefly motivate your choice(s).

#### Random Forest Classifier

Random forest is often used for classification, which is fitting for the current model (classification of blood cells). This method is based on ensemble learning, it combines  multiple classifiers and creates several decision trees. It goes through these trees and decides the best based on majority voting, then it delivers the output/prediction.

```python
# sklearn cannot handle the images, it need the data to be reformatted to a 2d array
# Training data
n_samples, n_x, n_y, n_rgb = X_train.shape
x_train2 = X_train.reshape((n_samples,n_x*n_y*n_rgb))
# Testing data
n_samples, n_x, n_y, n_rgb = X_test.shape
x_test2 = X_test.reshape((n_samples,n_x*n_y*n_rgb))
```

```python
print('Training shape: ', x_train2.shape)
print('Testing shape: ', x_test2.shape)
```

```python
# Create and fit model
model = RandomForestClassifier()
model.fit(x_train2 , y_train)
# Make predictions using model
y_pred = model.predict(x_test2)
```

```python
accuracy_score(y_pred,y_test)
print(classification_report(y_pred,y_test))
```

The random forest classifier seems to perform reasonably well on this subset with a weighted average usually around 70% (changes per iteration due to random state).

#### Decision Tree

Decision Tree is a supervised machine learning technique also often used for classification. It also works with creating trees, where each node ask a yes/no about a specific feature and further splits the trees. This way it is able to find patterns and classify something based on answers from the decision tree.

```python
dtc = DecisionTreeClassifier()
dtc.fit(x_train2, y_train)
y_pred_dtc = dtc.predict(x_test2)
```

```python
accuracy_score(y_pred_dtc,y_test)
print(classification_report(y_pred_dtc,y_test))
```

The decision tree classifier seems to perform quite poorly with this dataset, usually having a weighted average of around 30% (changes per iteration due to random state). In this case that would be slighly better then random guessing, however still a poor fit.

### Fully-connected neural network model

Design a fully-connected neural network that is able to perform the chosen classification/regression task. Properly motivate your design choices, or determine suitable values for model hyperparameters. Explain in particular how you decided on or arrived at: the depth of the model and the width of the various layers, the employed activation and loss functions, the batch size and number of epochs, and the employed optimizer. Show the model layout summary, train it on the available data, and reliably evaluate its performance. Determine whether in your opinion the model is over- and/or underfitting, and critically comment on how well your final model does in comparison to the baseline model(s), or potentially other model configurations that you tried. Reflect on potential further ways to improve the model that you did not have the time or opportunity to explore.

```python
# better explanation of the model available after plotting accuracy
neural = keras.models.Sequential()

# Add input layer with the size of each image/numpy array
neural.add(keras.layers.InputLayer(input_shape=(120, 160, 3), name='Input'))
neural.add(keras.layers.Flatten())  # flatten to give to dense layer
# Use three layers initially just to get a rough estimate whether more are needed.
# Adding more layers might result in just nothing happening, suggesting they aren't necessary
neural.add(keras.layers.Dense(units=128, activation='LeakyReLU', name='Hidden_1'))
neural.add(keras.layers.Dense(units=64, activation='LeakyReLU', name='Hidden_2'))
neural.add(keras.layers.Dense(units=32, activation='LeakyReLU', name='Hidden_3'))
# Add a dropout layer to combat possible overfitting
neural.add(keras.layers.Dropout(0.3))
# create output layer with 4 units (4 labels) and 
# softmax activation (normalized output adquate for multiple labels) 
neural.add(keras.layers.Dense(units=4, activation='softmax', name='Output'))

# generate summary for model
neural.summary()
```

Using categories, so "CategoricalCrossentropy" and "CategoricalAccuracy" are appropriate for loss and metrics. For gradient descend, Adam is one of the best optimizers despite being more computationally heavy. It adapts its movement during the process to optimize finding the minimum. When [compared to various optimizers ](https://towardsdatascience.com/optimizers-for-training-neural-network-59450d71caf6), Adam seems to be one that makes sure the local minima is not missed by overshooting.

```python
neural.compile(
    loss=keras.losses.CategoricalCrossentropy(),
    optimizer=keras.optimizers.Adam(),
    metrics=keras.metrics.CategoricalAccuracy()
)
```

```python
# epochs and batch size selected on trial and error to see what performs better
history = neural.fit(
    x=X_train,
    y=y_train,
    epochs=30,
    batch_size=250,
    verbose=1,
    validation_data=(X_test, y_test)
)

history.history.keys()
```

```python
plt.plot(history.history['loss'], 'go', label='Training loss')
plt.plot(history.history['val_loss'], 'g-', label='Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['categorical_accuracy'], 'go', label='Categorical accuracy')
plt.plot(history.history['val_categorical_accuracy'], 'g-', label='Validation categorical accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

The loss seems to decrease and quickly reach a minimum, however looking at the accuracy it is quite chaotic and clearly starts overfitting at around 17 epochs, with 26% accuracy. Considering the baseline models, this model performs quite poorly like the decision tree.

#### Observations

When it comes to neural networks, trial and error are often the first steps depending on the problem. Based on that output, parameters are adjusted.

Several parameters were tried with this fully connected model, the number of layers was varied from 3-5 but did not seem to alter the results. Afterwards different activation functions were tested: relu, leaky relu and tanh, once again no changes in results. Units varied in multiples of 2 down from 256 to 16. Despite every change, the model would not get much better then 30% accuracy and 25% validation accuracy. Even dropouts were performed after different layers in different combinations to see if at least the sligh overfitting at the end could be corrected but it did not change anything.

```python
neural.compile(
    loss=keras.losses.CategoricalCrossentropy(),
    optimizer=keras.optimizers.Adam(),
    metrics=keras.metrics.CategoricalAccuracy()
)
```
